### Act 9 Proyecto de simulación de inteligencia artificial  utiliza Lista,con if, else, elif  while tipo Menu
# Clase : Ingeniero

```python
# =====================================================================
#                 SISTEMA IA: Lista de Ingenieros
# =====================================================================

# Al inicio, la lista de conocimientos/registros (ahora también ingenieros) empieza vacía
ingenieros = []

print("=====================================================")
print("🤖 ¡Bienvenido al simulador de IA para Ingenieros! 🤖")
print("=====================================================")

ejecutando = True

# Ciclo principal usando 'while'
while ejecutando:
    print("\n--- MENÚ DE CONTROL DE LA IA ---")
    print("1. Registrar un nuevo ingeniero") # Nueva opción 1
    print("2. Diagnosticar estado de un proyecto (Tomar decisiones)") # Antes opción 1
    print("3. Enseñar un nuevo concepto a la IA") # Antes opción 2
    print("4. Mostrar la base de conocimientos y registros") # Antes opción 3
    print("5. Apagar la IA (Salir)") # Antes opción 4

    opcion = input("\nSelecciona una opción (1-5): ")

    # --- NUEVA OPCIÓN 1: REGISTRAR INGENIERO ---
    if opcion == "1":
        print("\n--- MÓDULO DE REGISTRO DE INGENIEROS ---")
        nombre_ingeniero = input("Escribe el nombre del ingeniero a registrar: ")
        ingenieros.append(f"Ingeniero registrado: {nombre_ingeniero}") # Añadir como string diferenciado
        print(f"-> IA: ¡Ingeniero '{nombre_ingeniero}' registrado en la lista 'ingenieros'!")

    # --- OPCIÓN 2 (ANTES OPCIÓN 1): DIAGNÓSTICO (Uso de IF, ELIF, ELSE) ---
    elif opcion == "2":
        print("\n--- MÓDULO DE DIAGNÓSTICO DE INGENIERÍA ---")
        print("¿Qué tipo de problema vas a analizar?")
        print("A) Estructural (Puentes/Edificios)")
        print("B) Eléctrico (Sistemas de energía)")

        tipo = input("Elige una opción (A o B): ").upper()

        if tipo == "A":
            grietas = input("¿El presupuesto está sobrepasado? (S/N): ").upper()
            if grietas == "S":
                print("-> IA: Alerta crítica. Recomiendo pausar la obra y recalcular costos.")
                ingenieros.append("Diagnóstico: Alerta de presupuesto en obra estructural.")
            else:
                print("-> IA: Todo en orden. Estructura y finanzas estables.")
                ingenieros.append("Diagnóstico: Proyecto estructural en condiciones óptimas.")

        elif tipo == "B":
            voltaje = input("¿El voltaje supera los 220V? (S/N): ").upper()
            if voltaje == "S":
                print("-> IA: Peligro de sobrecarga. Se sugiere instalar transformadores de seguridad.")
                ingenieros.append("Diagnóstico: Riesgo de sobrecarga eléctrica detectado.")
            else:
                print("-> IA: Flujo eléctrico normal y seguro.")
                ingenieros.append("Diagnóstico: Sistema eléctrico estable.")
        else:
            print("-> IA: Tipo de ingeniería no reconocido.")

    # --- OPCIÓN 3 (ANTES OPCIÓN 2): ENSEÑAR CONCEPTO A LA IA ---
    elif opcion == "3":
        print("\n--- MÓDULO DE APRENDIZAJE ---")
        nuevo_dato = input("Escribe una nueva regla o término de ingeniería para mi memoria: ")
        ingenieros.append(f"Conocimiento IA: {nuevo_dato}") # Añadir como string diferenciado
        print("-> IA: ¡Entendido! He guardado ese conocimiento en mi lista.")

    # --- OPCIÓN 4 (ANTES OPCIÓN 3): MOSTRAR LA LISTA USANDO WHILE (NO FOR) ---
    elif opcion == "4":
        print("\n--- MEMORIA ACTUAL DE LA IA y REGISTROS ---")
        # Validamos si la lista está vacía con un if-else
        if len(ingenieros) == 0:
            print("La lista está vacía. ¡Prueba las opciones 1, 2 o 3 primero!")
        else:
            print("Datos registrados:")
            # Recorremos la lista con un ciclo while e índice
            contador = 0
            while contador < len(ingenieros):
                print(f" Registro [{contador + 1}]: {ingenieros[contador]}")
                contador = contador + 1

    # --- OPCIÓN 5 (ANTES OPCIÓN 4): SALIR ---
    elif opcion == "5":
        print("\nIA: Guardando datos y cerrando sistemas... ¡Hasta la próxima, Ingeniero!")
        ejecutando = False

    # --- VALIDACIÓN DE OPCIÓN INCORRECTA ---
    else:
        print("Opción no válida. Por favor, elige un número del 1 al 5.")

print("\n=====================================================")
print("          Programa finalizado con éxito.")
print("=====================================================")
print("programa realizado por David Barrera 1388")
```
